# Graph theory / Network neuroscience
The brain consists of networks of brain regions all communicating with each other to varying extents. Network neuroscience refers to how researchers identity these discrete networks and their functional roles. 

Perhaps the most famous brain network is the default mode network, thought to be responsible for functioning when we are awake but not engaged in any specific mental task, specifically the network of brain areas which all decrease their activity during attention-demanding tasks and increase their activity when doing nothing. Such a network was only possible to detect given whole-brain data and large-scale explorations of brain activity, in contrast to more traditional ROI-specific analysis techniques. 

Although it can be possible to study brain networks without graph theory, graph theory is a useful tool for more nuanced detection of brain networks and how the networks functions. In graph theory, a network is a collection of nodes (or vertices) and their corresponding connections (or edges). The nodes are typically large brain regions (specifically "parcellations" of multiple brain areas that share some functional role), but graph theory is not limited to large-scale organization (nodes can be individual neurons) nor limited to functional networks (although structural networks may use different graph theory methods). 

Graph theory unfortunately contains a lot of jargony terms that may seem confusing at first. A "community" or "module" is simply a discrete network (mostly independent set of connected nodes) in a graph (where a "graph" is the full picture of all possible nodes/edges). "Neighboring nodes" simply refer to nodes that are directly connected (i.e., can reach the node with a single edge). "Cluster" refers to a group of nodes that are close together in a graph, and measures of clustering refer to a network is connected. A "hub" is a node that acts as a connector / transit point between communities. A "path" is a single possible route between nodes. "Degree" refers to the number of connections to a node. "Centrality" refers to how important certain nodes are in a graph (where "importance" can be defined in various ways, such as number of connections). "Rich club organization" refers to a group of highly interconnected nodes (i.e., more densely connected than would be expected by chance). "Small world networks" refer to networks where most nodes are not neighbors of one another, but for the nodes that are connected the average path length is very short.

# Simulation
First import the data and a few standard Python packages. We are also going to use a non-standard Python package called networkx which is specifically for visualizing networks.

```python
# general packages
import itertools
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# pull data from GitHub
import requests, io
for array in ['time_series']:
    globals()['{}'.format(array)] = np.load(io.BytesIO(requests.get(
        'http://github.com/PaulScotti/fmriplayground/raw/main/methods/graph_theory/{}.npy?raw=true'.format(array)).content),allow_pickle=True)

# change default plotting
plt.rcParams.update({'font.size': 16, 'figure.figsize': (16,4)})
    
# overview of the data
num_parcels = time_series.shape[0] #7
num_TRs = time_series.shape[1] #200
```

For simplicity, our simulated dataset consists of an averaged time course across participants for each of seven brain areas, where the participants were doing nothing in the scanner (i.e., resting-state). Our goal will be to visualize and quantify the network architecture of these brain areas.

In practice, this parcellation to group brain regions into large-scale areas would be derived from previous literature, grouping based on similar functioning. For instance, the below picture depicts the [Yeo et al. (2011)](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3174820/) 7-network coarse parcellation.


Yeo et al. (2011) parcellation into 7 brain areas

Let's first visualize our data, which is an averaged time course for each of the seven brain regions.

```python
for p in range(num_parcels):
    plt.plot(time_series[p,:],label='area {}'.format(p))
plt.xlabel("TR")
plt.ylabel("BOLD signal")
plt.legend()
plt.show()
```

We will now correlate every time course to every other time course to create a "connection matrix" which represents the functional relationships between brain areas. Note that other metrics besides correlation coefficient can be used in this way, and we are using correlation for simplicity.

```python
connection_matrix = np.full((num_parcels,num_parcels),1.0)
for p1,p2 in list(itertools.permutations(np.arange(num_parcels), 2)):
    connection_matrix[p1,p2] = np.corrcoef(time_series[p1,:],time_series[p2,:])[0][1]
    connection_matrix[p2,p1] = connection_matrix[p1,p2]

plt.imshow(connection_matrix,cmap="viridis")
plt.colorbar()
plt.title("Connection Matrix (R values)")
plt.show()
```

Now we will threshold this connection matrix to create an "adjacency matrix". This will remove the edges between nodes that are sufficiently weak to suggest they are not functionally connected. The choice of thresholding will vary depending on the specifics of the research question and the brain regions being explores, and so should be carefully considered alongside literature review. We will be using an arbitrary Pearson's R threshold of 0.05 for the sake of simplicity.

```python
adjacency_matrix = np.full((num_parcels,num_parcels),1.0)
for p1,p2 in list(itertools.permutations(np.arange(num_parcels), 2)):
    if connection_matrix[p1,p2] >= 0.05:
        adjacency_matrix[p1,p2] = connection_matrix[p1,p2]
    else:
        adjacency_matrix[p1,p2] = 0.0
    adjacency_matrix[p2,p1] = adjacency_matrix[p1,p2]

plt.imshow(adjacency_matrix,cmap="viridis")
plt.colorbar()
plt.title("Adjacency Matrix (R values)")
plt.show()
```

Now that we have an adjacency matrix, all we need to do is to visualize the network itself and compute network measures.

```python
# change default plotting
plt.rcParams.update({'font.size': 16, 'figure.figsize': (4,4)})

gr = nx.Graph()
for p in range(num_parcels):
    gr.add_node(p)
rows, cols = list((np.where(adjacency_matrix > 0)))
gr.add_edges_from(zip(rows, cols))
nx.draw(gr, node_size=400, node_color='lightblue', with_labels=True)
plt.show()
```

Great! Now we can compute all sorts of centrality measures as ways to quantify the topology of the above network. Below we will compute the degree distribution, which is the distribution of the number of connections between nodes.

```python
path_lengths = np.full(num_parcels,0)
for p1 in range(num_parcels):
    for p2 in range(num_parcels):
        if p1!=p2 and adjacency_matrix[p1,p2]>0:
            path_lengths[p1]+=1

plt.hist(path_lengths)
plt.ylabel("Frequency")
plt.xticks(np.arange(7))
plt.xlabel("# of connections")
plt.title("Degree distribution")
plt.show()
```

Now that we have some measures to quantify the dynamics of the network, the next step would typically involve comparison of measures (such as degree distribution, average path length, clustering coefficient, etc.) of the current network to other networks. For instance, you could compare to a random network to see how likely it would be for the observed degree distribution to be observed by random chance. Or you could compare the network of a clinical group to a network of a control group to see how the communication between brain areas may differ as a result of a clinical disorder.

```python
## Code used to simulate the data instead of loading it from GitHub: ##
import numpy as np
from scipy.stats import gamma
from scipy import signal
np.random.seed(1)
num_parcels = 7
num_TRs = 200

# trying to copy SPM canonical hrf with max value at 1
hrf_function = (gamma.pdf(np.arange(20),5) - 0.2 * gamma.pdf(np.arange(20),12)) / .2

def blocks(points,length=200):
    x = np.arange(length) # number of points on x axis
    y = np.zeros(length) # time series
    for p in points:
        y[p] = 1
    return y

time_series=np.full((num_parcels,num_TRs),np.nan)
for p in range(num_parcels):
    if p<num_parcels-1:
        steps = np.sort([np.random.randint(0, 190) for i in range(6)])
    else:
        steps[:4] = np.sort([np.random.randint(0, 190) for i in range(4)])
    block1 = blocks([np.arange(10)+steps[0],np.arange(10)+steps[1],np.arange(10)+steps[2],np.arange(10)+steps[3],
                     np.arange(10)+steps[4],np.arange(10)+steps[5]])
    block1_conv = (signal.convolve(block1, hrf_function, mode='full'))[:num_TRs]
    block1_conv /= np.max(block1_conv)
    
    time_series[p,:] = block1_conv+np.random.random(num_TRs)+np.random.random(num_TRs)

np.save('time_series.npy',time_series)
```
