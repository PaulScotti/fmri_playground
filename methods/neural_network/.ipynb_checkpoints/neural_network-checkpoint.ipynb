# Neural network
Note that "neural networks" are not at all related to actual "neurons"!

## Perceptron
A feedforward neural network with no hidden layers, also known as a single layer neural network (because there is a single link between one input layer and one output layer). A perceptron can perform linear regression or linear classification, whereas more complicated neural networks perform nonlinear operations. Perceptrons are basically identical to support vector machines in terms of functionality, with the exception that perceptrons seek to find any decision boundary that can separate classes whereas a support vector machine seeks to find the decision boundary that maximizes the distance between the classes. For this reason, support vector machines are generally preferable to perceptrons. Perceptrons are still very useful for educational purposes because they represent the most basic kind of a neural network.

Diagram of how perceptrons work in the case of voxel data. The left side of the diagram depicts the input layer, where each node is a unique voxel. The right side depicts the output layer consisting of a single node outputting whether the presented stimulus was 0 or 1. Each voxel's activation is multiplied by a voxel-specific weight. All these products are then summed and then added by a bias term. The result can be immediately used (for regression problems) or be fed into an activation function (e.g., step function) for classification problems. The predictions are compared to the ground truth and if there is a difference then the weights and bias are adjusted.

# Simulation: Perceptron
First import the data and a few standard Python packages.

```python
# general packages
import numpy as np
import matplotlib.pyplot as plt

# pull data from GitHub
import requests, io
for array in ['activations','stimulus']:
    globals()['{}'.format(array)] = np.load(io.BytesIO(requests.get(
        'http://github.com/PaulScotti/fmriplayground/raw/main/methods/neural_network/{}.npy?raw=true'.format(array)).content),allow_pickle=True)

# change default plotting
plt.rcParams.update({'font.size': 20, 'figure.figsize': (5,5)})
    
# overview of the data
num_conditions = len(np.unique(stimulus)) #2
num_trials = activations.shape[0] #50
num_voxels = activations.shape[1] #30
```

We have 1 subject who underwent 50 trials, where each trial was either watching a comedy movie clip (0) or a horror movie clip (1). We have activations for each trial for each of 30 voxels in our brain region of interest. Let's take a look at the overall histogram of activations for each condition.

```python
plt.hist(activations[stimulus==0,:].flatten(),color='blue',bins=30,label='Comedy')
plt.hist(activations[stimulus==1,:].flatten(),color='red',bins=30,alpha=.5,label='Horror')
plt.yticks([]);plt.ylabel('Frequency')
plt.xlabel("Activation strength")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

Lucky for us, the data appears to be linearly separable! Generally speaking, horror movie clips were associated with overall lower activation across voxels. We will train a perceptron to identify this observation and learn a set of weights that can be used to predict whether a given trial's voxel activations are associated with a comedy clip or a horror clip.

```python
# set model's random initial variables
weights = np.zeros(num_voxels)
bias = 0
learning_rate = 0.1
num_steps = 10

# setup example of first trial of training loop
trial = 0
print("voxel activations: {}".format(activations[trial,:]))
print("weights: {}".format(weights))
print("bias: {}".format(bias))
print("stimulus for trial 1: {}".format(stimulus[trial]))

# training first trial
summation = (activations[trial,:] @ weights) + bias
if summation > 0:
    prediction = 1
else:
    prediction = 0  
weights += learning_rate * (stimulus[trial] - prediction) * activations[trial,:]
bias += learning_rate * (stimulus[trial] - prediction)

print("\n")
print("prediction for trial 1: {}".format(prediction))
print("\n")
print("prediction incorrect, thus adjusting weights and bias (based on learning rate):")
print("weights: {}".format(weights))
print("bias: {}".format(bias))
```

The above code demonstrates the model attempting to predict the first trial's stimulus based on the initial weights and bias terms (here they were all set to zero). Given that the model is untrained, performance should be at chance level. Indeed, the prediction for the first trial was wrong. Given that the model was incorrect, the weights and bias are adjusted based on the learning rate (here arbitrarily set to 0.1). Specifically, the weights are increased by the learning rate multiplied by the voxel activations, and the bias term is increased by the learning rate. The intention here is that by changing the weights and bias slightly whenever error is encountered (and not adjusting the weights/bias if the prediction is correct) then the model will eventually move towards a state that is increasingly more accurate in making predictions.

```python
# rng seed makes results reproduceable
np.random.seed(1)

# reset model's random initial variables
weights = np.zeros(num_voxels)
bias = 0
predictions = np.full(num_trials,np.nan)
learning_rate = 0.1
num_steps = 10

# train model on first half of trials
for step in range(num_steps): # repeat training across all trials num_steps times
    for trial in range(num_trials//2):
        summation = (activations[trial,:] @ weights) + bias
        if summation > 0:
            activation = 1
        else:
            activation = 0  
        predictions[trial] = activation
        weights += learning_rate * (stimulus[trial] - predictions[trial]) * activations[trial,:]
        bias += learning_rate * (stimulus[trial] - predictions[trial])
        
percent_correct = np.sum(predictions[:num_trials//2]==stimulus[:num_trials//2]) / (num_trials/2) * 100
print("Percent Correct: {}%".format(percent_correct))
```

Wow! 100% accuracy! But hold on, consider the possibility that this accuracy is high because the model is overfit. That is, we are training the model on the first 25 trials and then asking how well the model predicts the presented stimulus on each of those 25 trials. Therefore, this is not an accurate test of how good our model is at predicting comedy/horror from voxel activations. To assess this, we need to predict trials that were not used to train the model.

```python
# test the trained model on second half of trials
for trial in range(num_trials//2,num_trials):
    summation = (activations[trial,:] @ weights) + bias
    if summation > 0:
        activation = 1
    else:
        activation = 0  
    predictions[trial] = activation
        
percent_correct = np.sum(predictions[num_trials//2:]==stimulus[num_trials//2:]) / (num_trials/2) * 100
print("Percent Correct: {}%".format(percent_correct))
```

The model's accuracy is far worse on the held-out test data than the training data, as would be expected. However, accuracy appears to still be above chance (50%).

Remember that a perceptron is simply a neural network without any hidden layers separating the input and output. Deep neural networks add additional hidden layers between the input and output layers, such that there are many more weights and biases. 

In practice, gradient descent is used to update the weights/biases rather than simply adding an arbitrary learning rate value across all terms. Our example's simplicity in how weights/bias are updated can be detrimental because it does not matter if voxel 1 has a bad weight and voxel 2 has a good weight -- both the weights will be similarly changed when the prediction is incorrect.

To get around this problem, backpropagation is used to calculate the loss/error of each node with respect to all the other nodes, going backwards from output layer to input layer. By using calculus (i.e., partial derivatives) to calculate the loss of every node, one can then use gradient descent to update the weights where nodes with higher error rates experience the greatest changes and the nodes with lower error rates experience the least changes. The vast majority of neural network tools and packages  automatically perform differentiation to accomplish these steps with minimal effort from the researcher.

```python
## Code used to simulate the data instead of loading it from GitHub: ##
import numpy as np
import pymc3 as pm
np.random.seed(1)
num_subjects = 1
num_trials = 50
num_voxels = 30
data=[]

normal1 = pm.Normal.dist(mu=1.5,sigma=.5)
normal2 = pm.Normal.dist(mu=.5,sigma=.5)
unif = pm.Uniform.dist(lower=-3,upper=3)
cond1 = pm.Mixture.dist(w=[.8,.2], comp_dists=[normal1,unif])
cond2 = pm.Mixture.dist(w=[.8,.2], comp_dists=[normal2,unif])

activations=np.full((num_trials,num_voxels),np.nan)
stimulus=np.full(num_trials,np.nan)

for t in range(num_trials):
    stim = np.random.randint(2)
    for v in range(num_voxels):
        if stim==0:
            activations[t,v]=cond1.random(size=1)
        else:
            activations[t,v]=cond2.random(size=1)
    stimulus[t]=stim

np.save('activations.npy',activations)
np.save('stimulus.npy',stimulus)
```
